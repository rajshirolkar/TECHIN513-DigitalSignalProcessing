```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft
```

```python
# generate a simple signal
fs = 100 # smpling grequency
t = np.arange(0, 10, 1/fs) # time vector
f = 10 # frequency of the actual signal
x = np.sin(2*np.pi*f*t)

plt.figure()
plt.plot(t,x)
```

```python
X = np.fft.fft(x)

# plot original signal and the FFT on the same figure
plt.figure()
plt.subplot(2,1,1)
plt.plot(t, x)
plt.xlabel('Time Domain (s)')
plt.ylabel('Amplitude')
plt.title('Original Signal in time')

plt.subplot(2,1,2)
plt.plot(np.abs(X))
plt.xlabel('Time Domain (s)')
plt.ylabel('Amplitude')
plt.title('FFT of Original Signal in time')

plt.subplots_adjust(hspace=0.5)
```
